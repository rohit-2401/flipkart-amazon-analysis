```python
# after getting the column name from the user and storing it in mask we print the product name and retail, discounted price from both the dataset
```

```python
al = inner_join[inner_join['product_name'] == mask]
al1= al.drop(['uniq_id_x','crawl_timestamp_x','product_url_x','product_category_tree_x','pid_x','image_x','is_FK_Advantage_product_x','description_x','product_rating_x',
'overall_rating_x',
'brand_x',
'product_specifications_x',
'uniq_id_y',
'crawl_timestamp_y',
'product_url_y',
'product_category_tree_y',
'pid_y',
'image_y',
'is_FK_Advantage_product_y',
'description_y',
'product_rating_y',
'overall_rating_y',
'brand_y',
'product_specifications_y'], axis=1)
al1
```

```python
# we are filtering the value and removing NaN values and removing index 
# then we duplicate the name column to bring in the format to print flipkart and amazon details separately
# we then put the data in a table to print it out as mentioned
```

```python
al2 = al1.dropna()
blankIndex=[''] * len(al2)
al2.index=blankIndex
al3 = al2.copy()
al3['product_name_amazon']= al3['product_name']
al3.rename(columns={'product_name':'product_name_flipkart'},inplace=True)
cols = list(al3.columns.values)
print(cols)
al3 = al3[['product_name_flipkart', 'retail_price_x', 'discounted_price_x', 'product_name_amazon', 'retail_price_y', 'discounted_price_y']]
print(tabulate(al3, headers = 'keys', tablefmt = 'fancy_grid'))
al3
```
